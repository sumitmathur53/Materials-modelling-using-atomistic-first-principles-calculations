```python
def read_poscar(file_name):

    data = open(file_name, "r")
    

    lines = data.readlines()
    data.close()

    # store the scale
    scale = float(lines[1])
    print("scaling factor: ", scale,'\n')

    # storing arrays of positions in a list

    positions_arrays = []

    for i in range(3):

        # extracting the single array position
        a = []
        for j in lines[i+2].split():
            a.append(j)
        # append the positions of single array
        positions_arrays.append(a)
    

    # now we a list corresponding to the positions
    positions_arrays = np.array(positions_arrays)


    print("position vectors -> ",'\n',positions_arrays)

    a1 = [float(val) for val in lines[2].split()]
    a2 = [float(val) for val in lines[3].split()]
    a3 = [float(val) for val in lines[4].split()]

    pos_vec = np.array([a1,a2,a3])
    pos_vec = scale * pos_vec


    # reading the types of atoms that is on the line 6th 
    atoms_types = lines[5].split()
    print('\n',"types of atoms present in the system ", atoms_types)


    # for reading the number of atoms of particular type 
    # creating an empty dictionary   
    atoms_types_count = {}
    for i in atoms_types:      
        j=0  
        atoms_types_count[i] = (lines[6].split()[j])
        j+=1
    print('\n',"types and count of atoms ", atoms_types_count)

    list_atoms_count = []

    
    for i in lines[6].split():
        list_atoms_count.append(int(i))

    print('\n','list of count', list_atoms_count)

    list_atoms_count = np.array(list_atoms_count)
    total_atoms_present_in_system = sum(list_atoms_count)

    print('\n','total_atoms_present_in_system ->', total_atoms_present_in_system)


    # now checking which coordinates are in this system cartesian or direct system

    direct_or_cartesian_system = lines[7].split()[0]
    direct_or_cartesian_system = direct_or_cartesian_system.lower()

    print('\n','type of system present :-> ',direct_or_cartesian_system)


    # to store the coordinates make a list
    coordinates = []

    # to store the atoms in the cartesian system
    atoms = []

    curr_line = 8

    for i in range(len(atoms_types)):
        atoms_symbol = atoms_types[i]
        atoms_symbol_count = list_atoms_count[i]

        for j in range(atoms_symbol_count):
            coordi = [float(val) for val in lines[curr_line].split()]
            curr_line += 1
            coordinates.append(coordi)
            atoms.append(atoms_symbol)

    

    coordinates = np.array(coordinates)
    print(coordinates.shape)
    print(coordinates)

    atoms = np.array(atoms)
    print(atoms.shape)
    print(atoms)

    print(type(positions_arrays))
    print(type(coordinates))

    print(positions_arrays.shape)

    print(type(pos_vec))

    print(pos_vec)

    if "d" == direct_or_cartesian_system[0]:
        direct_system = coordinates
        cartesian_system = np.matmul(direct_system,pos_vec)
    else:
        cartesian_system = scale * coordinates
        direct_system = np.matmul(cartesian_system, np.linalg.inv(pos_vec))

    
    

    print('\n', " direct system =", direct_system)
    print('\n', " cartesian system =", cartesian_system)


    data= {'cartesian_system' : cartesian_system,
                 'direct_system': direct_system,
                 'position_vector' : pos_vec,
                 'atoms' : atoms
                 } 
    return data
```

```python
def write_xyz_file(file_name , data):

    file = open(file_name,"w")
    file.write("%d \n" % (len(data['cartesian_system'])))
    print(len(data['cartesian_system']))
    

    a1 = " ".join([str(x) for x in data['position_vector'][0]])
    a2 = " ".join([str(x) for x in data['position_vector'][1]])
    a3 = " ".join([str(x) for x in data['position_vector'][2]])

    print("Lattice=",'"',a1, a2, a3,'"')

    file.write('Lattice="%s %s %s"' % (a1,a2,a3))

    
    for i in range(len(data['cartesian_system'])):
        
        
        print(data['atoms'][i], data['cartesian_system'][i][0],
                                       data['cartesian_system'][i][1],
                                       data['cartesian_system'][i][2])
        
        file.write("%s %f %f %f \n" % (data['atoms'][i], data['cartesian_system'][i][0],
                                       data['cartesian_system'][i][1],
                                       data['cartesian_system'][i][2]))
    

    
```
