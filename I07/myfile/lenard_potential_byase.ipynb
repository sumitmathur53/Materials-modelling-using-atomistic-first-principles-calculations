# ase lenard calculation


```python
import ase
import ase.io
import numpy as np
```

```python
atoms = ase.io.read('Ar_POSCAR', format='vasp')
```

```python
print (atoms)
```

```python
from ase.calculators.lj import LennardJones

calculator = LennardJones(epsilon=1.67e-2,sigma=3.4,rc=3.5*3.4)
```

```python
atoms.set_calculator(calculator)
```

```python
lattice_vector = [a/10.0 for a in range(50,56)]
print ("lattice vector", lattice_vector)

energies = []

for a in lattice_vector:

    cell = [[a,0,0],[0,a,0],[0,0,a]]

    atoms.set_cell(cell)

    energies.append(atoms.get_potential_energy())


for a,e in zip(lattice_vector, energies):
    print(a,e)
```

```python
import matplotlib.pyplot as plt
plt.plot(lattice_vector, energies, 'ro-')
plt.xlabel('Lattice COnstant (A)')
plt.ylabel('Energy (AU)')
```
