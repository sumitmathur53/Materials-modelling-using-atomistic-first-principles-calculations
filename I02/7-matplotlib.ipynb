Matplotlib is a Python plotting library which produces publication quality figures in a variety of hardcopy formats and interactive environments across platforms.

### Simple plot:

```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4], [1, 4, 9, 25])
plt.ylabel('some numbers')
plt.show()
```

### Formatting the style of your plot:

For every x, y pair of arguments, there is an optional third argument which is the format string that indicates the color and line type of the plot. The default format string is 'b-', which is a solid blue line. For example, to plot the above with red circles, you would issue:

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro-', linewidth=1.0)
plt.xlim([0,6])
plt.ylim([0,20])
plt.show()
```

### log-axis and more formatting:

```python
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(1., 5., 0.2)

# red dashes, blue squares and green triangles
plt.semilogy(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')  # plotting 3 datasets at a time
plt.legend(["linear", "Square", "Cube"])                 # names of these datasets
plt.xlabel('X')
plt.ylabel('f(X)')
plt.grid()                                               # show the background grid
plt.show()
```

### Scatter plot from dictionary:

```python
data = {'x-data': np.arange(50),                       # array of numbers from 0 to 50
        'color-data': np.random.randint(0, 50, 50),        # 50 random ints between 0 and 50
       }
data['y-data'] = data['x-data'] + 10 * np.random.randn(50)  
data['size-data'] = np.random.randint(0,100,50)

plt.scatter('x-data', 'y-data', c='color-data', s='size-data', data=data)    # c is color, and s is size
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

### histograms:

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

n, bins, patches = plt.hist(x, 50, density=True, facecolor='b', alpha=0.8)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(50, .025, 'mu=100, sigma=15')
plt.xlim(40, 160)
plt.ylim(0, 0.03)
plt.grid(True)
plt.show()
```
